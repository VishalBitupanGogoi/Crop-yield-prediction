```python
df['Yield']=(df['production']/df['area'])
df.head()
```

```python
df = df.rename({'season_Whole Year':'season_Whole_Year'}, axis=1)
df.head()
```

```python
df = df.rename({'crop_Arhar/Tur':'crop_Arhar_Tur','crop_Beans & Mutter(Vegetable)':'crop_Beans_Mutter_Vegetable','crop_Black pepper':'crop_Black_pepper','crop_Bottle Gourd':'crop_Bottle_Gourd','crop_Castor seed':'crop_Castor_seed','crop_Citrus Fruit':'crop_Citrus_Fruit','crop_Cotton(lint)':'crop_Cotton_lint',
                'crop_Cowpea(Lobia)':'crop_Cowpea_Lobia','crop_Dry chillies':'crop_Dry_chillies','crop_Dry ginger':'crop_Dry_ginger','crop_Horse-gram':'crop_Horse_gram', 'crop_Moong(Green Gram)':'crop_Moong_Green_Gram','crop_Oilseeds total':'crop_Oilseeds_total','crop_Other  Rabi pulses':'crop_Other_Rabi_pulses',
                'crop_Other Fresh Fruits':'crop_Other_Fresh_Fruits','crop_Other Kharif pulses':'crop_Other_Kharif_pulses','crop_Other Vegetables':'crop_Other_Vegetables','crop_Peas  (vegetable)':'crop_Peas_vegetable','crop_Pome Fruit':'crop_Pome_Fruit','crop_Pome Granet':'crop_Pome_Granet','crop_Pulses total':'crop_Pulses_total',
                'crop_Rapeseed &Mustard':'crop_Rapeseed_Mustard','crop_Small millets':'crop_Small_millets','crop_Sweet potato':'crop_Sweet_potato', 'crop_other fibres':'crop_other_fibres','crop_other misc. pulses':'crop_other_misc_pulses', 'crop_other oilseeds':'crop_other_oilseeds', 'crop_Niger seed':'crop_Niger_seed'}, axis=1)
df.head()
```

```python
cols = list(df.columns.values) #Make a list of all of the columns in the df
cols.pop(cols.index('Yield'))
df = df[cols+['Yield']] 
df.head()
```

```python
df.columns
```

```python
X = df.drop(columns=['Yield'])
y = df['Yield']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_random = RandomForestRegressor()
```

```python
#Hyperparameters
n_estimators = [int(x) for x in np.linspace(start=100, stop=1200, num=12)] #selecting diff decision trees
print(n_estimators)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
#Randomized Search CV, helps to find best parameters

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
# max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
# Create the random grid
# taking in the form of key value pair
random_grid = {'n_estimators':n_estimators,
               'max_features':max_features,
               'max_depth':max_depth,
               'min_samples_split':min_samples_split,
               'min_samples_leaf':min_samples_leaf}
print(random_grid) #randomizedCv will select the best parameters from this
```

```python
# Use the random grid to search for best hyperparams
#First create the best model to tune
rf = RandomForestRegressor()
```

```python
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
rf_random.fit(X_train,y_train)
```

```python
#our model is now ready now we do the predictions
predictions = rf_random.predict(X_test)
```

```python
predictions
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,predictions)
score
```
